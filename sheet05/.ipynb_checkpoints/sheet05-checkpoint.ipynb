## 4 Linear regions of MLPs



- The goal of this exercise is to build and investigate the behavior of Multi-Layer Perceptrons (MLPs) 
- with ReLU activations in terms of the number of linear regions they create in the input space.
- We will implement two models using PyTorch: one shallow model with a single hidden layer and another deeper model.
- We will visualize the output of these MLPs and examine their gradient behavior over a 2D input space.


Part (a)

```python
import torch
import torch.nn as nn

# Set a seed for reproducibility
torch.manual_seed(42)

# Define the shallow model
class ShallowMLP(nn.Module):
    def __init__(self):
        super(ShallowMLP, self).__init__()
        self.hidden = nn.Linear(2, 20)  # 2 input features and 20 neurons in hidden layer
        self.output = nn.Linear(20, 1)  # Output is a scalar value

    def forward(self, x):
        x = torch.relu(self.hidden(x))  # ReLU activation on hidden layer
        x = self.output(x)  # No activation on output layer
        return x

# Instantiate the model
model = ShallowMLP()

# Calculate the number of parameters
num_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'The number of parameters in the model: {num_params}')
```

### **Part (a): Implement a Shallow Model with a Single Hidden Layer of 20 Neurons**


**Explanation:**
- We start by setting a seed (`torch.manual_seed(42)`) to ensure that the random initialization of weights is reproducible, which is important when trying to replicate the results.
- The model, named `ShallowMLP`, is a simple Multi-Layer Perceptron (MLP) with:
  - An **input layer** of size 2, which corresponds to the two-dimensional input (`x ∈ ℝ²`).
  - A **hidden layer** with **20 neurons**.
  - An **output layer** producing a scalar value (`y ∈ ℝ¹`).
  - The hidden layer uses **ReLU** activation (`torch.relu()`), while the output layer has no activation function.
- We then compute the **number of parameters** in the model:
  - **Input to Hidden Layer:** Weights (2 inputs × 20 neurons = 40) + Biases (20 biases) = **60**.
  - **Hidden to Output Layer:** Weights (20 neurons × 1 output = 20) + Bias (1 bias) = **21**.
  - The **total number of parameters** is **81**.


Part (b)

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility in generating input data
np.random.seed(42)

# Create a grid of points in the range [-10, 10] x [-10, 10]
x_range = np.linspace(-10, 10, 500)
y_range = np.linspace(-10, 10, 500)
xx, yy = np.meshgrid(x_range, y_range)
grid = np.c_[xx.ravel(), yy.ravel()]

# Convert grid to a tensor
grid_tensor = torch.tensor(grid, dtype=torch.float32)

# Pass grid through the model
with torch.no_grad():
    z = model(grid_tensor).numpy()

# Reshape the output to match the grid shape
z = z.reshape(xx.shape)

# Plot the output as an image
plt.imshow(z, extent=(-10, 10, -10, 10), origin='lower', cmap='viridis')
plt.colorbar()
plt.title("Network Output for Shallow Model")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### **Part (b): Generate an Image of the Network Output**


**Explanation:**
- We set a seed (`np.random.seed(42)`) to ensure that any random operations performed with NumPy are consistent between runs.
- A **grid of input points** is generated in the range `[-10, 10]` for both x and y, creating a 500 × 500 grid. This grid helps us visualize how the model responds to different inputs.
- The **grid of points** is passed through the shallow model, and the resulting outputs are visualized using a **heatmap** (`imshow()` from `matplotlib`).
- The output plot represents how the model's prediction changes across the input space. The `viridis` color map is used to show different values with distinct colors.
- **ReLU activation** causes regions with negative input to become zero, resulting in noticeable boundaries in the output.

Part (C)

```python
# Compute the gradient using numpy's gradient function
dz_dx, dz_dy = np.gradient(z, x_range, y_range)

# Plot the gradient components
plt.imshow(dz_dx, extent=(-10, 10, -10, 10), origin='lower', cmap='prism')
plt.colorbar()
plt.title("Gradient of Network Output (dz/dx)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

plt.imshow(dz_dy, extent=(-10, 10, -10, 10), origin='lower', cmap='prism')
plt.colorbar()
plt.title("Gradient of Network Output (dz/dy)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### **Part (c): Compute and Visualize the Spatial Gradient of Network Output**


**Explanation:**
- The **gradient** of the output is computed with respect to the input using `np.gradient()`. This function calculates numerical derivatives of the model output.
- The **spatial gradients** `dz/dx` and `dz/dy` are computed for the x and y directions, respectively. These gradients indicate how rapidly the model's output changes with respect to each input dimension.
- Each component (`dz/dx` and `dz/dy`) is visualized separately using `imshow()`.
- The **gradient visualization** shows the sensitivity of the model's output to changes in the x and y inputs.
  - Areas with sharper color transitions in the gradient plots indicate higher sensitivity.
  - The **piecewise linear structure** observed is a result of the ReLU activation, which creates sharp boundaries between regions where the output changes and where it stays constant.
- The **prism color map** is used here for better differentiation of gradient regions.


Part (d)

```python
# Set seed for reproducibility
torch.manual_seed(42)

# Define the deeper model with four hidden layers of 5 neurons each
class DeepMLP(nn.Module):
    def __init__(self):
        super(DeepMLP, self).__init__()
        self.layer1 = nn.Linear(2, 5)
        self.layer2 = nn.Linear(5, 5)
        self.layer3 = nn.Linear(5, 5)
        self.layer4 = nn.Linear(5, 5)
        self.output = nn.Linear(5, 1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.relu(self.layer2(x))
        x = torch.relu(self.layer3(x))
        x = torch.relu(self.layer4(x))
        x = self.output(x)
        return x

# Instantiate the model
deep_model = DeepMLP()

# Pass the grid through the deep model and visualize (similar to part b)
with torch.no_grad():
    z_deep = deep_model(grid_tensor).numpy()

# Reshape the output to match the grid shape
z_deep = z_deep.reshape(xx.shape)

# Plot the output as an image
plt.imshow(z_deep, extent=(-10, 10, -10, 10), origin='lower', cmap='viridis')
plt.colorbar()
plt.title("Network Output for Deep Model")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### **Part (d): Implement a Deeper Model with Four Hidden Layers of 5 Neurons Each**

**Explanation:**
- The **deeper model** (`DeepMLP`) is implemented with **four hidden layers**, each having **5 neurons**.
- The deeper model has more **expressive power** compared to the shallow model due to its greater depth, allowing it to represent more complex functions.
- We again use the **ReLU activation function** after each hidden layer, which introduces non-linearity, making the model capable of capturing complex relationships in the data.
- The output of the **deep model** is visualized similarly to the shallow model by generating a grid of input values.
- **Comparison to Shallow Model**:
  - The **shallow model** has simpler and smoother transitions in the output, representing fewer regions.
  - The **deep model** creates a more complex structure in the output, indicating that it can capture finer details and more intricate relationships in the data.
  - This additional complexity can help solve problems where the relationship between input and output is more non-linear, but it also means the model might be more prone to **overfitting** if not regularized appropriately.
