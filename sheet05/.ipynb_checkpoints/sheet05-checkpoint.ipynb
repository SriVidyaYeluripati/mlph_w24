## 4 Linear regions of MLPs



- The goal of this exercise is to build and investigate the behavior of Multi-Layer Perceptrons (MLPs) 
- with ReLU activations in terms of the number of linear regions they create in the input space.
- We will implement two models using PyTorch: one shallow model with a single hidden layer and another deeper model.
- We will visualize the output of these MLPs and examine their gradient behavior over a 2D input space.


```python
# Part (a): Shallow Model Implementation
# Import necessary libraries
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import random

# Step 1: Set the random seed for reproducibility
# Setting the seed ensures that our weight initialization and other random processes give consistent results
torch.manual_seed(42)
torch.cuda.manual_seed(42)
torch.cuda.manual_seed_all(42)  # If we use multiple GPUs
np.random.seed(42)
random.seed(42)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# Step 2: Define the Shallow MLP model
class ShallowMLP(nn.Module):
    def __init__(self):
        super(ShallowMLP, self).__init__()
        # Define a hidden layer with 20 neurons, and ReLU activation
        self.hidden = nn.Linear(2, 20)  # Input: 2 features, Output: 20 neurons in the hidden layer
        self.relu = nn.ReLU()           # Activation function after the hidden layer
        # Define the output layer, which gives a single output
        self.output = nn.Linear(20, 1)  # Input: 20 neurons, Output: 1 (scalar)

    def forward(self, x):
        # Forward pass through hidden layer and activation
        x = self.hidden(x)
        x = self.relu(x)
        # Forward pass through output layer
        x = self.output(x)
        return x

# Step 3: Initialize the Shallow MLP model
shallow_model = ShallowMLP()
print(f"Total number of parameters in the shallow model: {sum(p.numel() for p in shallow_model.parameters())}")

# Step 4: Generate a dense grid of 500x500 points in the range [-10, 10] for both input dimensions
x1 = np.linspace(-10, 10, 500)
x2 = np.linspace(-10, 10, 500)
# Create a meshgrid from these points
x1_grid, x2_grid = np.meshgrid(x1, x2)

# Step 5: Flatten the grid and prepare it for passing through the MLP
input_grid = np.vstack([x1_grid.ravel(), x2_grid.ravel()]).T
input_tensor = torch.tensor(input_grid, dtype=torch.float32)  # Convert to tensor for PyTorch

# Step 6: Pass the input grid through the shallow model and collect the output
# We use torch.no_grad() to prevent gradient calculation during inference
with torch.no_grad():
    output = shallow_model(input_tensor).numpy()  # Forward pass to get output

# Step 7: Reshape the output back into the 500x500 grid for visualization
output_reshaped = output.reshape(500, 500)

# Step 8: Visualize the output of the shallow model as an image
plt.imshow(output_reshaped, extent=[-10, 10, -10, 10], origin='lower', cmap='viridis')
plt.colorbar()
plt.title('Output of Shallow MLP over 2D Grid')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Part (b): Gradient Analysis of Shallow Model
# Step 9: Calculate the spatial gradient of the output using numpy.gradient
# Compute the gradient along both axes of the output_reshaped
gradient_x, gradient_y = np.gradient(output_reshaped)

# Step 10: Visualize the gradient components to understand the regions formed by the MLP
# Visualizing gradient w.r.t. x1
plt.imshow(gradient_x, extent=[-10, 10, -10, 10], origin='lower', cmap='prism')
plt.colorbar()
plt.title('Gradient of Output with respect to x1 (Shallow Model)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()

# Visualizing gradient w.r.t. x2
plt.imshow(gradient_y, extent=[-10, 10, -10, 10], origin='lower', cmap='prism')
plt.colorbar()
plt.title('Gradient of Output with respect to x2 (Shallow Model)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Part (c): Deep Model Implementation
# Step 11: Define a deeper MLP model with 4 hidden layers of 5 neurons each
class DeepMLP(nn.Module):
    def __init__(self):
        super(DeepMLP, self).__init__()
        # Define four hidden layers, each with 5 neurons
        self.layer1 = nn.Linear(2, 5)
        self.layer2 = nn.Linear(5, 5)
        self.layer3 = nn.Linear(5, 5)
        self.layer4 = nn.Linear(5, 5)
        # Define the output layer which produces a single scalar output
        self.output = nn.Linear(5, 1)
        self.relu = nn.ReLU()  # Activation function

    def forward(self, x):
        # Forward pass through all layers
        x = self.relu(self.layer1(x))
        x = self.relu(self.layer2(x))
        x = self.relu(self.layer3(x))
        x = self.relu(self.layer4(x))
        x = self.output(x)
        return x

# Step 12: Initialize the Deep MLP model
deep_model = DeepMLP()
print(f"Total number of parameters in the deep model: {sum(p.numel() for p in deep_model.parameters())}")

# Step 13: Pass the same input grid through the deep model and collect the output
with torch.no_grad():
    deep_output = deep_model(input_tensor).numpy()  # Forward pass to get output

# Step 14: Reshape the output of the deep model for visualization
deep_output_reshaped = deep_output.reshape(500, 500)

# Step 15: Visualize the output of the deep model as an image
plt.imshow(deep_output_reshaped, extent=[-10, 10, -10, 10], origin='lower', cmap='viridis')
plt.colorbar()
plt.title('Output of Deep MLP over 2D Grid')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Part (d): Gradient Analysis of Deep Model
# Step 16: Calculate the spatial gradient of the deep model's output
deep_gradient_x, deep_gradient_y = np.gradient(deep_output_reshaped)

# Step 17: Visualize the gradient components of the deep model's output
# Visualizing gradient w.r.t. x1 for deep model
plt.imshow(deep_gradient_x, extent=[-10, 10, -10, 10], origin='lower', cmap='prism')
plt.colorbar()
plt.title('Gradient of Output with respect to x1 (Deep Model)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()

# Visualizing gradient w.r.t. x2 for deep model
plt.imshow(deep_gradient_y, extent=[-10, 10, -10, 10], origin='lower', cmap='prism')
plt.colorbar()
plt.title('Gradient of Output with respect to x2 (Deep Model)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
